```python
# count plot
plt.title('species Count')
sns.countplot(x=iris['Species'])
```

```python
 #comparison between sepal length and sepal width
plt.title('Comparison between sepal width and length')
sns.scatterplot(x=iris['SepalLengthCm'], y=iris['SepalWidthCm'])
```

```python
# comparison between sepal length and sepal width on the basis of species
plt.title('Comparison between sepal length and width')
sns.scatterplot(x = iris['SepalLengthCm'], y = iris['SepalWidthCm'], hue = iris['Species'])
```

```python
# comparison between petal length and petal width
plt.title('Comparison between petal width and length')
sns.scatterplot(x = iris['PetalLengthCm'], y = iris['PetalWidthCm'])
```

```python
# comparison between petal length and petal width
plt.title('Comparison between petal width and length')
sns.scatterplot(x = iris['PetalLengthCm'], y = iris['PetalWidthCm'], hue = iris['Species'])
```

```python
sns.violinplot(x=iris['Species'],y=iris['SepalLengthCm'])
```

```python
#Plot 1
sns.FacetGrid(iris, hue="Species", height=5).map(sns.distplot, "SepalLengthCm").add_legend()

#Plot 2
sns.FacetGrid(iris, hue="Species", height=5) \
.map(sns.distplot, "SepalWidthCm") \
.add_legend()

#Plot 3
sns.FacetGrid(iris, hue="Species", height=5) \
.map(sns.distplot, "PetalLengthCm") \
.add_legend()

#Plot 4
sns.FacetGrid(iris, hue="Species", height=5) \
.map(sns.distplot, "PetalWidthCm") \
.add_legend()

plt.show()
```
